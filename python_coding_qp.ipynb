```python
# 1. Write a python function to reverse a given string?
def reverse_stg(inp):
    return inp[::-1]

input_str = "Hello, World!"
output_str = reverse_stg(input_str)
print(output_str)
```

```python
# 2. Check if a string is a palindrome?
def is_palindrome(word):
    word = word.lower()
    return word == word[::-1]

word = "level"
output_word = is_palindrome(word)
print(output_word)
```

```python
# 3. Find the first non-repeated character in a string?
def non_repeated_char(in_str):
    char_count = {}
    for char in in_str:
        char_count[char] = char_count.get(char, 0) + 1

    for char in in_str:
        if char_count[char] == 1:
            return char
    return None

    
in_str = "Programming"
out_char = non_repeated_char(in_str)
print(out_char)
```

```python
# 4. Calculate the factorial of number?
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)

input_num = 5
output_result = factorial(input_num)
print(output_result)
```

```python
# 5. Merge two sorted lists?
def merge_sorted_lists(l1, l2):
    merged_list = sorted(l1 + l2)
    return merged_list

l1 = [1, 3, 5]
l2 = [2, 4, 6]
output_list = merge_sorted_lists(l1, l2)
print(output_list)
```
